```python
time_dict = {'Jan':'01', 'Feb':'02', 'Mar':'03', 'Apr':'04', 'May':'05', 'Jun':'06', 
            'Jul':'07', 'Aug':'08', 'Sep':'09', 'Oct':'10', 'Nov':'11', 'Dec':'12'}
def Time_transfer(t):
    dd = t[4:6]
    if dd[1] == ',':
        dd = '0' + dd[0]
    mm = time_dict[t[:3]]
    year = t[-4:]
    return year + mm + dd
```
